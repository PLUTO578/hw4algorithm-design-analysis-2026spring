## A 排序

```python
## add your code here
import java.io.InputStream;
import java.io.IOException;
import java.util.ArrayList;
import java.util.Collections;

public class Main {
    static final int N = 1024;
    static int n, a, b, L;
    static int[] p = new int[N];
    static int[] pos = new int[N];
    static ArrayList<Integer> ops = new ArrayList<>();

    static void upd_pos() {
        for (int i = 0; i < n; ++i) pos[p[i]] = i;
    }

    static void swp() {
        ops.add(0);
        for (int i = 0; i < n; ++i) {
            if (p[i] == a) p[i] = b;
            else if (p[i] == b) p[i] = a;
        }
        upd_pos();
    }

    static void add(int x) {
        if (x == 0) return;
        ops.add(x);
        for (int i = 0; i < n; ++i) p[i] = (p[i] + x) % n;
        upd_pos();
    }

    static void xxr(int x) {
        if (x == 0) return;
        ops.add(-x);
        for (int i = 0; i < n; ++i) p[i] ^= x;
        upd_pos();
    }

    static int[] get_ps(int a, int b) {
        int d = (b - a + n - L + n) % n;
        int pa = 0, pb = 0;
        for (int s = n / 2; s >= 2 * L; s /= 2) {
            if (d >= s) {
                d -= s;
                pb += s / 2;
            } else {
                pa += s / 2;
            }
        }
        pa += n / 2 + (a & (L - 1));
        pb += (a & (L - 1));
        return new int[]{pa, pb};
    }

    static void swap_el(int c, int d) {
        if ((c / L) % 2 == (d / L) % 2) {
            int t;
            if ((c / L) % 2 == 0) t = (c & (L - 1)) + L;
            else t = (c & (L - 1));
            swap_el(c, t);
            swap_el(d, t);
            swap_el(c, t);
        } else {
            int[] resAB = get_ps(a, b);
            int pa = resAB[0], pb = resAB[1];

            int[] resCD = get_ps(c, d);
            int pc = resCD[0], pd = resCD[1];

            add((pc - c + n) % n);
            xxr(pc ^ pa);
            add((a - pa + n) % n);
            swp();
            add((pa - a + n) % n);
            xxr(pc ^ pa);
            add((c - pc + n) % n);
        }
    }

    static class P {
        int[] a = new int[N];
        int sz;
        ArrayList<Integer> v = new ArrayList<>();

        boolean build() {
            boolean[] vis = new boolean[N];
            for (int i = 0; i < sz; ++i) {
                if (a[i] >= 0 && a[i] < N) {
                    vis[a[i]] = true;
                }
            }
            for (int i = 0; i < sz; ++i) if (!vis[i]) return false;

            if (sz == 1) return true;

            P ev = new P();
            P od = new P();
            for (int i = 0; i < sz / 2; ++i) {
                ev.a[i] = a[2 * i] / 2;
                od.a[i] = a[2 * i + 1] / 2;
            }
            ev.sz = od.sz = sz / 2;

            if (!ev.build() || !od.build()) return false;
            if (a[0] % 2 != 0) v.add(sz == 2 ? 1 : -1);

            int te = 0, to = 0;
            for (int x : ev.v) {
                if (x > 0) {
                    v.add(-1); v.add(1);
                } else {
                    v.add(2 * x); te ^= -2 * x;
                }
            }
            if (te != 0) v.add(-te);

            for (int x : od.v) {
                if (x > 0) {
                    v.add(1); v.add(-1);
                } else {
                    v.add(2 * x); to ^= -2 * x;
                }
            }

            if ((to & (sz / 2)) != (te & (sz / 2))) {
                for (int i = 0; i < sz / 4; ++i) {
                    v.add(-1); v.add(1);
                }
            }
            if (te >= sz / 2) te -= sz / 2;
            if (to >= sz / 2) to -= sz / 2;
            if (te != to) return false;

            ArrayList<Integer> tmp = new ArrayList<>();
            for (int x : v) {
                if (tmp.isEmpty()) {
                    tmp.add(x);
                } else if (x < 0 && tmp.get(tmp.size() - 1) < 0) {
                    int last = tmp.get(tmp.size() - 1);
                    int newVal = -((-last) ^ (-x));
                    tmp.set(tmp.size() - 1, newVal);
                    if (newVal == 0) tmp.remove(tmp.size() - 1);
                } else {
                    tmp.add(x);
                }
            }
            v = tmp;
            return true;
        }
    }

    static class FastReader {
        InputStream is;
        public FastReader(InputStream is) {
            this.is = is;
        }
        int nextInt() {
            try {
                int c = is.read();
                while (c < '0' || c > '9') {
                    if (c == -1) return -1;
                    c = is.read();
                }
                int res = 0;
                while (c >= '0' && c <= '9') {
                    res = res * 10 + (c - '0');
                    c = is.read();
                }
                return res;
            } catch (IOException e) {
                return -1;
            }
        }
    }

    public static void main(String[] args) {
        FastReader sc = new FastReader(System.in);
        n = sc.nextInt();
        if (n == -1) return;
        a = sc.nextInt();
        b = sc.nextInt();
        for (int i = 0; i < n; ++i) {
            p[i] = sc.nextInt();
        }
        upd_pos();

        L = (a - b + n) % n;
        L = L & -L;
        if (L == 0) L = n;

        if (L > 1) {
            P sub = new P();
            sub.sz = L;
            for (int i = 0; i < n; ++i) sub.a[i] = p[i] & (L - 1);
            if (!sub.build()) {
                System.out.println("-1");
                System.exit(0);
            }
            for (int x : sub.v) {
                if (x > 0) add(x);
                else xxr(-x);
            }
        }

        for (int r = 0; r < L; ++r) {
            ArrayList<Integer> tmp = new ArrayList<>();
            for (int j = r; j < n; j += L) tmp.add(p[j]);
            Collections.sort(tmp);

            int idx = 0;
            boolean ok = true;
            for (int j = r; j < n; j += L) {
                if (tmp.get(idx) != j) { ok = false; break; }
                idx++;
            }
            if (!ok) {
                System.out.println("-1");
                System.exit(0);
            }
            for (int j = r; j < n; j += L) {
                if (p[j] != j) swap_el(j, p[j]);
            }
        }

        for (int i = 0; i < n; ++i) {
            if (p[i] != i) {
                throw new RuntimeException("Assertion failed: Element mismatch");
            }
        }

        System.out.println(ops.size());
        StringBuilder sb = new StringBuilder();
        for (int x : ops) {
            if (x == 0) {
                sb.append("0\n");
            } else if (x < 0) {
                sb.append("1 ").append(-x).append("\n");
            } else {
                sb.append("2 ").append(x).append("\n");
            }
        }
        System.out.print(sb);
    }
}
```

## B 长跑

```python
## add your code here
#include <iostream>
#include <vector>
#include <algorithm>

using namespace std;

struct Station {
    int p, c;

    bool operator<(const Station& other) const {
        if (p != other.p) return p < other.p;
        return c < other.c;
    }
};

void solve() {
    int N, L, Maxn, S;
    while (cin >> N >> L >> Maxn >> S) {
        vector<Station> stations;
        for (int i = 0; i < N; ++i) {
            int p, c;
            cin >> p >> c;
            if (c <= S && p <= L) {
                stations.push_back({p, c});
            }
        }

        // 1. 0次补给就能到终点的情况
        if (Maxn >= L) {
            cout << "Yes\n";
            continue;
        }

        sort(stations.begin(), stations.end());

        vector<Station> unique_stations;
        for (const auto& st : stations) {
            if (unique_stations.empty() || unique_stations.back().p != st.p) {
                unique_stations.push_back(st);
            }
        }

        bool possible = false;

        // 2. 检查只需 1 次补给的情况
        for (const auto& st : unique_stations) {
            if (st.p <= Maxn && L - st.p <= Maxn && st.c <= S) {
                possible = true;
                break;
            }
        }

        if (possible) {
            cout << "Yes\n";
            continue;
        }

        // 3. 检查需要 2 次补给的情况
        int m = unique_stations.size();
        for (int i = 0; i < m; ++i) {
            if (unique_stations[i].p > Maxn) break;

            for (int j = i + 1; j < m; ++j) {
                if (unique_stations[j].p - unique_stations[i].p > Maxn) break;

                if (L - unique_stations[j].p <= Maxn) {
                    if (unique_stations[i].c + unique_stations[j].c <= S) {
                        possible = true;
                        break;
                    }
                }
            }
            if (possible) break;
        }

        if (possible) {
            cout << "Yes\n";
        } else {
            cout << "No\n";
        }
    }
}

int main() {
    ios_base::sync_with_stdio(false);
    cin.tie(NULL);

    solve();

    return 0;
}
```

## C 最长回文

```python
## add your code here
import java.io.*;

public class Main {

    static class StringHash {
        long[] h1, h2;
        long[] p1, p2;
        int n;
        static final long M1 = 1000000007L;
        static final long M2 = 1000000009L;
        static final long B1 = 131;
        static final long B2 = 13331;

        public StringHash(String s) {
            n = s.length();
            h1 = new long[n + 1];
            h2 = new long[n + 1];
            p1 = new long[n + 1];
            p2 = new long[n + 1];
            p1[0] = 1;
            p2[0] = 1;
            for (int i = 1; i <= n; i++) {
                p1[i] = (p1[i - 1] * B1) % M1;
                p2[i] = (p2[i - 1] * B2) % M2;
                h1[i] = (h1[i - 1] * B1 + s.charAt(i - 1)) % M1;
                h2[i] = (h2[i - 1] * B2 + s.charAt(i - 1)) % M2;
            }
        }

        public long getHash(int l, int r) {
            long hash1 = (h1[r] - h1[l - 1] * p1[r - l + 1] % M1 + M1) % M1;
            long hash2 = (h2[r] - h2[l - 1] * p2[r - l + 1] % M2 + M2) % M2;
            return (hash1 << 32) | hash2;
        }
    }

    public static int[] manacher(String s) {
        int n = s.length();
        char[] t = new char[2 * n + 3];
        t[0] = '$';
        t[1] = '#';
        for (int i = 0; i < n; i++) {
            t[2 * i + 2] = s.charAt(i);
            t[2 * i + 3] = '#';
        }
        t[2 * n + 2] = '@';

        int[] p = new int[t.length];
        int center = 0, right = 0;
        for (int i = 1; i < t.length - 1; i++) {
            if (i < right) {
                p[i] = Math.min(right - i, p[2 * center - i]);
            }
            while (t[i + p[i] + 1] == t[i - p[i] - 1]) {
                p[i]++;
            }
            if (i + p[i] > right) {
                center = i;
                right = i + p[i];
            }
        }
        return p;
    }

    public static void main(String[] args) throws IOException {
        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));
        String nStr = br.readLine();
        if (nStr == null) return;

        int n = Integer.parseInt(nStr.trim());
        String A = br.readLine().trim();
        String B = br.readLine().trim();

        String A_rev = new StringBuilder(A).reverse().toString();

        StringHash hashA_rev = new StringHash(A_rev);
        StringHash hashB = new StringHash(B);

        int[] pA = manacher(A);
        int[] pB = manacher(B);

        int maxLen = 0;

        for (int i = 2; i <= 2 * n; i++) {
            int p = pA[i];
            if (p == 0) continue;
            int l = (i - p + 1) / 2;
            int r = (i + p - 1) / 2;
            if (l <= r) {
                int x = l - 1;
                int y = r;
                int lcp = getLCP(x, y, n, hashA_rev, hashB);
                int len = (r - l + 1) + 2 * lcp;
                maxLen = Math.max(maxLen, len);
            }
        }

        for (int i = 2; i <= 2 * n; i++) {
            int p = pB[i];
            if (p == 0) continue;
            int l = (i - p + 1) / 2;
            int r = (i + p - 1) / 2;
            if (l <= r) {
                int x = l;
                int y = r + 1;
                int lcp = getLCP(x, y, n, hashA_rev, hashB);
                int len = (r - l + 1) + 2 * lcp;
                maxLen = Math.max(maxLen, len);
            }
        }

        for (int i = 1; i <= n; i++) {
            int x = i;
            int y = i;
            int lcp = getLCP(x, y, n, hashA_rev, hashB);
            int len = 2 * lcp;
            maxLen = Math.max(maxLen, len);
        }

        System.out.println(maxLen);
    }

    static int getLCP(int x, int y, int n, StringHash hashA_rev, StringHash hashB) {
        if (x < 1 || y > n) return 0;
        int low = 0;
        int high = Math.min(x, n - y + 1);
        int ans = 0;

        while (low <= high) {
            int mid = (low + high) / 2;
            if (mid == 0) {
                ans = Math.max(ans, 0);
                low = mid + 1;
                continue;
            }
            long hA = hashA_rev.getHash(n - x + 1, n - x + mid);
            long hB = hashB.getHash(y, y + mid - 1);
            if (hA == hB) {
                ans = mid;
                low = mid + 1;
            } else {
                high = mid - 1;
            }
        }
        return ans;
    }
}
```

## D 优惠券

```python
## add your code here
import java.io.BufferedReader;
import java.io.InputStreamReader;
import java.io.IOException;
import java.util.StringTokenizer;

public class Main {
    static final int MAX_X = 100005;
    static int[] last_pos_arr = new int[MAX_X];
    static int[] state_arr = new int[MAX_X];

    static class Op {
        int type;
        int x;
        Op(int type, int x) {
            this.type = type;
            this.x = x;
        }
    }

    static int findSet(int i, int[] parent) {
        if (parent[i] == i) return i;
        return parent[i] = findSet(parent[i], parent);
    }

    public static void main(String[] args) throws IOException {
        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));
        StringBuilder out = new StringBuilder();
        String line;

        while ((line = br.readLine()) != null) {
            line = line.trim();
            if (line.isEmpty()) continue;

            int m = Integer.parseInt(line);
            if (m == 0) {
                out.append("-1\n");
                continue;
            }

            Op[] ops = new Op[m + 1];
            int[] parent = new int[m + 2];

            for (int i = 1; i <= m; ++i) {
                StringTokenizer st = new StringTokenizer(br.readLine());
                String s = st.nextToken();

                if (s.charAt(0) == 'I') {
                    int x = Integer.parseInt(st.nextToken());
                    ops[i] = new Op(1, x);
                    parent[i] = i + 1;
                } else if (s.charAt(0) == 'O') {
                    int x = Integer.parseInt(st.nextToken());
                    ops[i] = new Op(2, x);
                    parent[i] = i + 1;
                } else {
                    ops[i] = new Op(0, 0);
                    parent[i] = i;
                }
            }
            parent[m + 1] = m + 1;

            int[] touched = new int[m + 1];
            int touched_count = 0;

            int error_line = -1;

            for (int i = 1; i <= m; ++i) {
                if (ops[i].type == 0) continue;

                int type = ops[i].type;
                int x = ops[i].x;
                touched[touched_count++] = x;

                if (type == 1) {
                    if (state_arr[x] == 1) {
                        int p = findSet(last_pos_arr[x] + 1, parent);
                        if (p < i) {
                            parent[p] = p + 1;
                            last_pos_arr[x] = i;
                        } else {
                            error_line = i;
                            break;
                        }
                    } else {
                        state_arr[x] = 1;
                        last_pos_arr[x] = i;
                    }
                } else if (type == 2) {
                    if (state_arr[x] == 0) {
                        int p = findSet(last_pos_arr[x] + 1, parent);
                        if (p < i) {
                            parent[p] = p + 1;
                            last_pos_arr[x] = i;
                        } else {
                            error_line = i;
                            break;
                        }
                    } else {
                        state_arr[x] = 0;
                        last_pos_arr[x] = i;
                    }
                }
            }

            if (error_line != -1) {
                out.append(error_line).append("\n");
            } else {
                out.append("-1\n");
            }

            for (int i = 0; i < touched_count; ++i) {
                int x = touched[i];
                last_pos_arr[x] = 0;
                state_arr[x] = 0;
            }
        }

        System.out.print(out);
    }
}
```

```python
## add your code here
import java.util.*;

public class Main {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);

        if (!sc.hasNextInt()) return;
        int n = sc.nextInt();

        int[] x = new int[n];
        int[] y = new int[n];

        for (int i = 0; i < n; i++) {
            x[i] = sc.nextInt();
            y[i] = sc.nextInt();
        }

        boolean[] visited = new boolean[n];
        int components = 0;

        for (int i = 0; i < n; i++) {
            if (!visited[i]) {
                components++;
                dfs(i, n, x, y, visited);
            }
        }

        System.out.println(components - 1);
    }

    private static void dfs(int curr, int n, int[] x, int[] y, boolean[] visited) {
        visited[curr] = true;
        for (int i = 0; i < n; i++) {
            if (!visited[i]) {
                if (x[curr] == x[i] || y[curr] == y[i]) {
                    dfs(i, n, x, y, visited);
                }
            }
        }
    }
}
```

## F 通配符匹配

```python
## add your code here
import java.io.BufferedReader;
import java.io.InputStreamReader;
import java.io.IOException;
import java.util.ArrayList;
import java.util.List;

public class Main {
    static final int BASE = 131;
    static final int MOD = 1_000_000_009;
    static long[] POW;

    static class Block {
        String text;
        long hash;
        int[] qPos;

        public Block(String text) {
            this.text = text;
            long h = 0;
            int count = 0;
            for (int i = 0; i < text.length(); i++) {
                if (text.charAt(i) == '?') count++;
            }
            qPos = new int[count];
            int idx = 0;
            for (int i = 0; i < text.length(); i++) {
                char c = text.charAt(i);
                if (c == '?') {
                    qPos[idx++] = i;
                    h = (h * BASE) % MOD;
                } else {
                    h = (h * BASE + c) % MOD;
                }
            }
            this.hash = h;
        }
    }

    static boolean match(String S, String B, int sIdx) {
        if (sIdx + B.length() > S.length()) return false;
        for (int i = 0; i < B.length(); i++) {
            char bc = B.charAt(i);
            if (bc != '?' && bc != S.charAt(sIdx + i)) {
                return false;
            }
        }
        return true;
    }

    static long getHash(int L, int R, long[] h) {
        long val = (h[R + 1] - h[L] * POW[R - L + 1]) % MOD;
        return (val < 0) ? val + MOD : val;
    }

    static int findMatch(String S, Block B, int sIdx, int rightLimit, long[] h) {
        int len = B.text.length();
        if (len == 0) return sIdx;

        for (int i = sIdx; i <= rightLimit - len; i++) {
            long curHash = getHash(i, i + len - 1, h);

            for (int p : B.qPos) {
                char sc = S.charAt(i + p);
                long subtract = (sc * POW[len - 1 - p]) % MOD;
                curHash = (curHash - subtract + MOD) % MOD;
            }

            if (curHash == B.hash) {
                if (match(S, B.text, i)) {
                    return i;
                }
            }
        }
        return -1;
    }

    public static void main(String[] args) throws IOException {
        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));
        String P = br.readLine();
        if (P == null) return;
        P = P.trim();

        String nStr = br.readLine();
        while (nStr != null && nStr.trim().isEmpty()) nStr = br.readLine();
        if (nStr == null) return;
        int n = Integer.parseInt(nStr.trim());

        POW = new long[100005];
        POW[0] = 1;
        for (int i = 1; i <= 100000; i++) {
            POW[i] = (POW[i - 1] * BASE) % MOD;
        }

        if (!P.contains("*")) {
            for (int i = 0; i < n; i++) {
                String S = br.readLine().trim();
                if (S.length() == P.length() && match(S, P, 0)) {
                    System.out.println("YES");
                } else {
                    System.out.println("NO");
                }
            }
            return;
        }

        List<Block> blocks = new ArrayList<>();
        StringBuilder sb = new StringBuilder();
        for (int i = 0; i < P.length(); i++) {
            char c = P.charAt(i);
            if (c == '*') {
                if (sb.length() > 0) {
                    blocks.add(new Block(sb.toString()));
                    sb.setLength(0);
                }
            } else {
                sb.append(c);
            }
        }
        if (sb.length() > 0) {
            blocks.add(new Block(sb.toString()));
        }

        boolean hasLeadingStar = P.charAt(0) == '*';
        boolean hasTrailingStar = P.charAt(P.length() - 1) == '*';

        StringBuilder out = new StringBuilder();

        for (int fileIdx = 0; fileIdx < n; fileIdx++) {
            String S = br.readLine().trim();

            if (blocks.isEmpty()) { // 若模式串全是 '*'
                out.append("YES\n");
                continue;
            }

            long[] h = new long[S.length() + 1];
            for (int i = 0; i < S.length(); i++) {
                h[i + 1] = (h[i] * BASE + S.charAt(i)) % MOD;
            }

            int sIdx = 0;
            int rightLimit = S.length();
            int startBlock = 0;
            int endBlock = blocks.size() - 1;
            boolean possible = true;

            if (!hasLeadingStar) {
                Block B0 = blocks.get(0);
                if (S.length() < B0.text.length() || !match(S, B0.text, 0)) {
                    possible = false;
                } else {
                    sIdx = B0.text.length();
                    startBlock = 1;
                }
            }

            if (possible && !hasTrailingStar && startBlock <= endBlock) {
                Block B_last = blocks.get(endBlock);
                int matchPos = S.length() - B_last.text.length();
                if (matchPos < sIdx || !match(S, B_last.text, matchPos)) {
                    possible = false;
                } else {
                    rightLimit = matchPos;
                    endBlock--;
                }
            }

            if (possible) {
                for (int i = startBlock; i <= endBlock; i++) {
                    Block B = blocks.get(i);
                    int matchPos = findMatch(S, B, sIdx, rightLimit, h);
                    if (matchPos == -1) {
                        possible = false;
                        break;
                    }
                    sIdx = matchPos + B.text.length();
                }
            }

            out.append(possible ? "YES\n" : "NO\n");
        }
        System.out.print(out.toString());
    }
}
```

## G 汉诺塔

```python
## add your code here
import java.util.*;

public class Main {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);
        int n = sc.nextInt();
        sc.nextLine();
        String[] priority = sc.nextLine().split(" ");

        long[][] steps = new long[n + 1][3];
        int[][] target = new int[n + 1][3];

        for (int i = 0; i < 3; i++) {
            char startChar = (char) ('A' + i);
            for (String op : priority) {
                if (op.charAt(0) == startChar) {
                    steps[1][i] = 1;
                    target[1][i] = op.charAt(1) - 'A';
                    break;
                }
            }
        }

        for (int k = 2; k <= n; k++) {
            for (int s = 0; s < 3; s++) {
                int t1 = target[k - 1][s];
                int t2 = 3 - s - t1;

                int t3 = target[k - 1][t1];

                if (t3 == t2) {
                    target[k][s] = t2;
                    steps[k][s] = steps[k - 1][s] + 1 + steps[k - 1][t1];
                } else {
                    target[k][s] = t1;
                    steps[k][s] = steps[k - 1][s] + 1 + steps[k - 1][t1] + 1 + steps[k - 1][s];
                }
            }
        }

        System.out.println(steps[n][0]);
    }
}
```

## H 马步距离

```python
## add your code here
import java.util.Scanner;

public class Main {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);

        if (!sc.hasNextLong()) return;
        long x1 = sc.nextLong();
        long y1 = sc.nextLong();
        long x2 = sc.nextLong();
        long y2 = sc.nextLong();

        long dx = Math.abs(x1 - x2);
        long dy = Math.abs(y1 - y2);

        if (dx < dy) {
            long temp = dx;
            dx = dy;
            dy = temp;
        }

        long ans;
        if (dx == 1 && dy == 0) {
            ans = 3;
        } else if (dx == 2 && dy == 2) {
            ans = 4;
        } else {
            ans = Math.max((dx + 1) / 2, (dx + dy + 2) / 3);

            while (ans % 2 != (dx + dy) % 2) {
                ans++;
            }
        }

        System.out.println(ans);
    }
}
```

## I 直方图最大矩形

```python
## add your code here
import java.util.Stack;

public class Solution {
    public int largestRectangleArea (int[] heights) {
        if (heights == null || heights.length == 0) {
            return 0;
        }

        Stack<Integer> stack = new Stack<>();
        int maxArea = 0;
        int n = heights.length;

        for (int i = 0; i <= n; i++) {
            int currentHeight = (i == n) ? 0 : heights[i];

            while (!stack.isEmpty() && currentHeight < heights[stack.peek()]) {
                int height = heights[stack.pop()];

                int width = stack.isEmpty() ? i : i - stack.peek() - 1;

                maxArea = Math.max(maxArea, height * width);
            }

            stack.push(i);
        }

        return maxArea;
    }
}
```

## J 消防局的设立

```python
## add your code here
import java.io.BufferedReader;
import java.io.IOException;
import java.io.InputStreamReader;
import java.util.ArrayList;
import java.util.Arrays;
import java.util.LinkedList;
import java.util.List;
import java.util.Queue;

public class Main {
    public static void main(String[] args) throws IOException {
        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));
        String line = br.readLine();
        if (line == null || line.trim().isEmpty()) return;

        int n = Integer.parseInt(line.trim());

        List<Integer>[] adj = new ArrayList[n + 1];
        for (int i = 1; i <= n; i++) {
            adj[i] = new ArrayList<>();
        }

        for (int i = 2; i <= n; i++) {
            line = br.readLine();
            if (line == null) break;
            int u = Integer.parseInt(line.trim());
            adj[i].add(u);
            adj[u].add(i);
        }

        int[] parent = new int[n + 1];
        int[] order = new int[n + 1];
        int[] dist = new int[n + 1];
        Arrays.fill(dist, 1000000);

        Queue<Integer> queue = new LinkedList<>();
        queue.add(1);
        parent[1] = 0;
        int idx = 1;

        while (!queue.isEmpty()) {
            int u = queue.poll();
            order[idx++] = u;
            for (int v : adj[u]) {
                if (v != parent[u]) {
                    parent[v] = u;
                    queue.add(v);
                }
            }
        }

        int ans = 0;

        for (int i = n; i >= 1; i--) {
            int u = order[i];

            if (dist[u] > 2) {
                int p = parent[u];
                int gp = parent[p];

                if (p == 0) gp = u;
                else if (gp == 0) gp = p;

                ans++;

                dist[gp] = 0;
                for (int v : adj[gp]) {
                    dist[v] = Math.min(dist[v], 1);
                    for (int w : adj[v]) {
                        dist[w] = Math.min(dist[w], 2);
                    }
                }
            }
        }

        System.out.println(ans);
    }
}
```
